```python
import os
import torchaudio
import torch
from torch.utils.data import Dataset, DataLoader

class ASRDataset(Dataset):
    def __init__(self, wav_dir, txt_dir, tokenizer, transform=None):
        self.wav_dir = wav_dir
        self.txt_dir = txt_dir
        self.transform = transform
        self.tokenizer = tokenizer

        self.files = [f[:-4] for f in os.listdir(wav_dir) if f.endswith('.wav')]
        # Ses ve metin dosyasının ikisi de var mı kontrol edip devam edelim
        self.files = [f for f in self.files if os.path.exists(os.path.join(txt_dir, f + '.txt'))]

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        fname = self.files[idx]
        wav_path = os.path.join(self.wav_dir, fname + '.wav')
        txt_path = os.path.join(self.txt_dir, fname + '.txt')

        waveform, sample_rate = torchaudio.load(wav_path)
        if self.transform:
            waveform = self.transform(waveform)

        with open(txt_path, 'r', encoding='utf-8') as f:
            transcript = f.read().strip().lower()

        # Tokenize transcript
        transcript_tokens = self.tokenizer.encode(transcript)

        return waveform, transcript_tokens

# Basit bir tokenizer (Türkçe karakter bazlı)
class CharTokenizer:
    def __init__(self):
        self.chars = list(" abcçdefgğhıijklmnoöprsştuüvyz'.,!?-")
        self.char2idx = {c: i + 1 for i, c in enumerate(self.chars)}
        self.idx2char = {i + 1: c for i, c in enumerate(self.chars)}

    def encode(self, text):
        return [self.char2idx.get(c, 0) for c in text]

    def decode(self, tokens):
        return ''.join([self.idx2char.get(t, '') for t in tokens])

# Örnek kullanımı
wav_dir = './wav_16k_mono'
txt_dir = './'
tokenizer = CharTokenizer()

dataset = ASRDataset(wav_dir, txt_dir, tokenizer)
print(f"Dataset size: {len(dataset)}")
waveform, tokens = dataset[0]
print(waveform.shape, tokens)
```

```python
import torch
from torch.nn.utils.rnn import pad_sequence
from torch.utils.data import DataLoader

def collate_fn(batch):
    """
    batch: list of tuples (waveform, token_list)
    waveform: tensor [1, time]
    token_list: list of int (tokens)
    """
    waveforms = [item[0].squeeze(0).t() for item in batch]
    tokens = [torch.tensor(item[1], dtype=torch.long) for item in batch]

    # waveforms padding
    waveforms_padded = pad_sequence(waveforms, batch_first=True)

    # tokens padding
    tokens_padded = pad_sequence(tokens, batch_first=True, padding_value=0)

    # Masklar
    wave_lengths = torch.tensor([w.shape[0] for w in waveforms], dtype=torch.long)
    token_lengths = torch.tensor([t.shape[0] for t in tokens], dtype=torch.long)

    return waveforms_padded, wave_lengths, tokens_padded, token_lengths

# DataLoader oluşturma
batch_size = 4
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True, collate_fn=collate_fn)

# DataLoader test
for batch in dataloader:
    waveforms_padded, wave_lengths, tokens_padded, token_lengths = batch
    print(f"Waveforms batch shape: {waveforms_padded.shape}")
    print(f"Wave lengths: {wave_lengths}")
    print(f"Tokens batch shape: {tokens_padded.shape}")
    print(f"Token lengths: {token_lengths}")
    break  # sadece ilk batch'i gösterelim
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Model Tanımı
class SimpleASRModel(nn.Module):
    def __init__(self):
        super(SimpleASRModel, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv1d(1, 32, kernel_size=5, stride=2, padding=2),
            nn.ReLU(),
            nn.Conv1d(32, 64, kernel_size=5, stride=2, padding=2),
            nn.ReLU(),
            nn.Conv1d(64, 128, kernel_size=5, stride=2, padding=2),
            nn.ReLU(),
        )
        self.lstm = nn.LSTM(128, 256, num_layers=2, batch_first=True, bidirectional=True)
        self.classifier = nn.Linear(512, 37)  # 37 karakter (örnek için)

    def forward(self, x):
        # x: [batch, 1, time]
        x = self.conv(x)              # [batch, channel, time']
        x = x.permute(0, 2, 1)        # [batch, time', channel]
        x, _ = self.lstm(x)           # [batch, time', hidden*2]
        x = self.classifier(x)        # [batch, time', classes]
        return x

# Conv çıkış uzunluğunu hesaplayan yardımcı fonksiyon
def get_output_length(input_length):
    for _ in range(3):
        input_length = (input_length + 2*2 - 5) // 2 + 1
    return input_length

# Model, optimizer ve loss
model = SimpleASRModel().to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.CTCLoss(blank=0, zero_infinity=True)

# Eğitim parametreleri
num_epochs = 20


for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for batch in dataloader:
        waveforms, wave_lengths, tokens, token_lengths = batch

        waveforms = waveforms.to(device)
        tokens = tokens.to(device)
        wave_lengths = wave_lengths.to(device)
        token_lengths = token_lengths.to(device)

        # 🔧 Çok kanallı sesleri mono'ya indirelim
        if waveforms.ndim == 3 and waveforms.shape[1] > 1:
            waveforms = waveforms.mean(dim=1, keepdim=True)
        elif waveforms.ndim == 2:
            waveforms = waveforms.unsqueeze(1)

        # İleri geçiş
        optimizer.zero_grad()
        outputs = model(waveforms)
        outputs = F.log_softmax(outputs, dim=2)

        input_lengths = torch.tensor(
            [get_output_length(l.item()) for l in wave_lengths],
            dtype=torch.long
        ).to(device)

        # CTC Loss
        loss = criterion(outputs.permute(1, 0, 2), tokens, input_lengths, token_lengths)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    avg_loss = running_loss / len(dataloader)
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {avg_loss:.4f}")
```

```python
# Karakter sözlüğü örneği
vocab = ['<blank>'] + list("abcdefghijklmnopqrstuvwxyz '")

def greedy_decoder(output, vocab):
    """
    output: [batch, time, num_classes] — log_softmax çıktı
    returns: list of decoded strings
    """
    argmaxes = output.argmax(dim=2)
    decoded_batch = []

    for indices in argmaxes:
        decoded = []
        prev = -1
        for idx in indices:
            idx = idx.item()
            if idx != prev and idx != 0:
                decoded.append(vocab[idx])
            prev = idx
        decoded_batch.append(''.join(decoded))
    return decoded_batch
```

```python
import torch.nn as nn
import torch.nn.functional as F

class SimpleASRModel(nn.Module):
    def __init__(self, num_classes):
        super(SimpleASRModel, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv1d(1, 32, kernel_size=5, stride=2, padding=2),
            nn.ReLU(),
            nn.Conv1d(32, 64, kernel_size=5, stride=2, padding=2),
            nn.ReLU(),
            nn.Conv1d(64, 128, kernel_size=5, stride=2, padding=2),
            nn.ReLU(),
        )
        self.lstm = nn.LSTM(128, 256, num_layers=2, batch_first=True, bidirectional=True)
        self.classifier = nn.Linear(512, num_classes)  # karakter sayısı kadar

    def forward(self, x):
        x = self.conv(x)
        x = x.permute(0, 2, 1)
        x, _ = self.lstm(x)
        x = self.classifier(x)
        return x
```

```python
import torch.optim as optim

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

num_classes = len(tokenizer.chars)
model = SimpleASRModel(num_classes).to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.CTCLoss(blank=0, zero_infinity=True)

def get_output_length(input_length):
    for _ in range(3):  # conv katmanlarının output length hesaplaması
        input_length = (input_length + 2*2 - 5) // 2 + 1
    return input_length

num_epochs = 10

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for batch in dataloader:
        waveforms, wave_lengths, tokens, token_lengths = batch

        waveforms = waveforms.to(device)
        tokens = tokens.to(device)
        wave_lengths = wave_lengths.to(device)
        token_lengths = token_lengths.to(device)

        # Eğer stereo ise mono yapalım
        if waveforms.ndim == 3 and waveforms.shape[1] > 1:
            waveforms = waveforms.mean(dim=1, keepdim=True)
        elif waveforms.ndim == 2:
            waveforms = waveforms.unsqueeze(1)

        optimizer.zero_grad()
        outputs = model(waveforms)
        outputs = F.log_softmax(outputs, dim=2)

        input_lengths = torch.tensor(
            [get_output_length(l.item()) for l in wave_lengths],
            dtype=torch.long
        ).to(device)

        loss = criterion(outputs.permute(1,0,2), tokens, input_lengths, token_lengths)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    avg_loss = running_loss / len(dataloader)
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {avg_loss:.4f}")
```

```python
class CharTokenizer:
    def __init__(self):
        self.chars = ['<pad>', '<s>', '</s>'] + list(" abcçdefgğhıijklmnoöprsştuüvyz.,!?")
        self.char2idx = {ch: idx for idx, ch in enumerate(self.chars)}
        self.idx2char = {idx: ch for ch, idx in self.char2idx.items()}

    def encode(self, text):
        return [self.char2idx['<s>']] + [self.char2idx.get(c, 0) for c in text] + [self.char2idx['</s>']]

    def decode(self, idxs):
        return ''.join([self.idx2char.get(i, '') for i in idxs if i > 2])
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from torch.nn.utils.rnn import pad_sequence

# Dataset
class PunctuationDataset(Dataset):
    def __init__(self, text_pairs, tokenizer):
        self.pairs = text_pairs
        self.tokenizer = tokenizer

    def __len__(self):
        return len(self.pairs)

    def __getitem__(self, idx):
        src_text, tgt_text = self.pairs[idx]
        src_ids = torch.tensor(self.tokenizer.encode(src_text), dtype=torch.long)
        tgt_ids = torch.tensor(self.tokenizer.encode(tgt_text), dtype=torch.long)
        return src_ids, tgt_ids


def collate_fn(batch):
    src_batch, tgt_batch = zip(*batch)
    src_batch = pad_sequence(src_batch, batch_first=True, padding_value=0)
    tgt_batch = pad_sequence(tgt_batch, batch_first=True, padding_value=0)
    return src_batch, tgt_batch
```

```python
import torch.nn as nn

#Model
class TinyTransformer(nn.Module):
    def __init__(self, vocab_size, embed_dim=128, hidden_dim=256, num_layers=2, num_heads=4):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        encoder_layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=num_heads, batch_first=True)
        self.encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.fc = nn.Linear(embed_dim, vocab_size)

    def forward(self, x):
        embedded = self.embedding(x)
        encoded = self.encoder(embedded)
        output = self.fc(encoded)
        return output
```

```python
import torch.nn.functional as F
import torch.optim as optim

def train(model, dataloader, optimizer, criterion, device):
    model.train()
    total_loss = 0

    for src, tgt in dataloader:
        src, tgt = src.to(device), tgt.to(device)

        optimizer.zero_grad()
        output = model(src)

        # output ve tgt şekillerini eşitlemek için seq_len'i bulalım
        seq_len = min(output.size(1), tgt.size(1))

        output = output[:, :seq_len, :]
        tgt = tgt[:, :seq_len]

        output = output.reshape(-1, output.size(-1))
        tgt = tgt.reshape(-1)

        loss = criterion(output, tgt)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(dataloader)

```

```python
import os
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torch.optim as optim
import re

#Basit tokenizer
class SimpleTokenizer:
    def __init__(self, chars=None):
        if chars is None:
            chars = " abcçdefgğhıijklmnoöprsştuüvyz0123456789.,!?-"
        self.chars = sorted(set(chars))
        self.char2idx = {c: i for i, c in enumerate(self.chars)}
        self.idx2char = {i: c for i, c in enumerate(self.chars)}

    def encode(self, text):
        return [self.char2idx.get(c, 0) for c in text.lower()]

    def decode(self, indices):
        return ''.join(self.idx2char.get(i, '') for i in indices)

#Dataset
class SceneDataset(Dataset):
    def __init__(self, scene_texts, labels, tokenizer, max_len=200):
        self.scene_texts = scene_texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.scene_texts)

    def __getitem__(self, idx):
        text = self.scene_texts[idx]
        label = self.labels[idx]
        encoded = self.tokenizer.encode(text)[:self.max_len]

        padded = encoded + [0] * (self.max_len - len(encoded))
        return torch.tensor(padded), torch.tensor(label)

#Transformer
class TransformerClassifier(nn.Module):
    def __init__(self, vocab_size, embed_dim=64, num_heads=4, num_layers=2, num_classes=10, max_len=200):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        encoder_layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=num_heads)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.fc = nn.Linear(embed_dim * max_len, num_classes)

    def forward(self, x):
        x = self.embedding(x)
        x = x.permute(1, 0, 2)
        x = self.transformer(x)
        x = x.permute(1, 0, 2)
        x = x.reshape(x.size(0), -1)
        x = self.fc(x)
        return x

#Eğitim Fonksiyonu
def train(model, dataloader, optimizer, criterion, device):
    model.train()
    total_loss = 0
    for inputs, labels in dataloader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(dataloader)


#Sahne metinleri ve sahne numaraları
scene_texts = [
    "Bir çocuk ormanda yürüyordu ve kuşlar şarkı söylüyordu.",
    "Ama sonra karanlık çöktü ve çocuk korkmaya başladı.",
    "Sonra sabah oldu ve güneş tekrar parladı."
]
labels = [0, 1, 2]  #sahne numaraları


tokenizer = SimpleTokenizer()
dataset = SceneDataset(scene_texts, labels, tokenizer)
dataloader = DataLoader(dataset, batch_size=2, shuffle=True)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = TransformerClassifier(vocab_size=len(tokenizer.chars), num_classes=3).to(device)

optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()

#Eğitim
epochs = 10
for epoch in range(epochs):
    loss = train(model, dataloader, optimizer, criterion, device)
    print(f"Epoch {epoch+1}/{epochs}, Loss: {loss:.4f}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np

#CPU olarak sabitliyorum
device = torch.device("cpu")

#Basit Generator
class Generator(nn.Module):
    def __init__(self, z_dim=100, img_channels=3, img_size=32):
        super().__init__()
        self.gen = nn.Sequential(
            nn.Linear(z_dim, 256),
            nn.ReLU(True),
            nn.Linear(256, 512),
            nn.ReLU(True),
            nn.Linear(512, 1024),
            nn.ReLU(True),
            nn.Linear(1024, img_channels * img_size * img_size),
            nn.Tanh()
        )
        self.img_channels = img_channels
        self.img_size = img_size

    def forward(self, z):
        x = self.gen(z)
        x = x.view(-1, self.img_channels, self.img_size, self.img_size)
        return x

#Basit Discriminator
class Discriminator(nn.Module):
    def __init__(self, img_channels=3, img_size=32):
        super().__init__()
        self.disc = nn.Sequential(
            nn.Linear(img_channels * img_size * img_size, 512),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )
        self.img_channels = img_channels
        self.img_size = img_size

    def forward(self, img):
        x = img.view(-1, self.img_channels * self.img_size * self.img_size)
        x = self.disc(x)
        return x

#Hiperparametreler
z_dim = 100
batch_size = 64
lr = 0.0002
epochs = 20
img_size = 32
img_channels = 3

#(CIFAR10)
transform = transforms.Compose([
    transforms.Resize(img_size),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)  # -1 ile 1 arası
])

dataset = datasets.CIFAR10(root="./data", train=True, download=True, transform=transform)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

#Model ve Optimizasyon
generator = Generator(z_dim, img_channels, img_size).to(device)
discriminator = Discriminator(img_channels, img_size).to(device)

criterion = nn.BCELoss()
opt_gen = optim.Adam(generator.parameters(), lr=lr)
opt_disc = optim.Adam(discriminator.parameters(), lr=lr)

#Eğitim Döngüsü
for epoch in range(epochs):
    for batch_idx, (real, _) in enumerate(dataloader):
        real = real.to(device)
        batch_size_curr = real.size(0)

        # Gerçek ve sahte etiketleri tanımla
        real_labels = torch.ones(batch_size_curr, 1).to(device)
        fake_labels = torch.zeros(batch_size_curr, 1).to(device)

        #Discriminator Eğitimi
        outputs = discriminator(real)
        d_loss_real = criterion(outputs, real_labels)
        real_score = outputs

        z = torch.randn(batch_size_curr, z_dim).to(device)
        fake_images = generator(z)
        outputs = discriminator(fake_images.detach())
        d_loss_fake = criterion(outputs, fake_labels)
        fake_score = outputs

        d_loss = d_loss_real + d_loss_fake
        discriminator.zero_grad()
        d_loss.backward()
        opt_disc.step()

        #Generator Eğitimi
        z = torch.randn(batch_size_curr, z_dim).to(device)
        fake_images = generator(z)
        outputs = discriminator(fake_images)
        g_loss = criterion(outputs, real_labels)  # Gerçek gibi görünmesini istiyorum

        generator.zero_grad()
        g_loss.backward()
        opt_gen.step()

    print(f"Epoch [{epoch+1}/{epochs}] | D Loss: {d_loss.item():.4f} | G Loss: {g_loss.item():.4f}")

#Örnek Görüntü Oluşturma
def show_generated_images(gen, num_images=16):
    gen.eval()
    z = torch.randn(num_images, z_dim).to(device)
    with torch.no_grad():
        fake_images = gen(z).cpu()
    fake_images = fake_images * 0.5 + 0.5  # Normalize [0,1]
    grid = np.transpose(torchvision.utils.make_grid(fake_images, nrow=4), (1,2,0))
    plt.figure(figsize=(6,6))
    plt.imshow(grid)
    plt.axis('off')
    plt.show()

show_generated_images(generator)
```

```python

df = pd.read_csv('/content/drive/MyDrive/sahne_betimlemeleri_labeled.csv')

# Train/val olarak bölelim
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

#Tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

#Dataset ve DataLoader
train_dataset = SceneDataset(train_df, tokenizer)
val_dataset = SceneDataset(val_df, tokenizer)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8)
```

```python
class SceneDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_len=64):
        self.data = dataframe
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        text = self.data.iloc[idx]['scene']
        label = self.data.iloc[idx]['label']
        encoded = self.tokenizer(text, padding='max_length', truncation=True, max_length=self.max_len, return_tensors='pt')
        return {
            'input_ids': encoded['input_ids'].squeeze(0),
            'attention_mask': encoded['attention_mask'].squeeze(0),
            'labels': torch.tensor(label)
        }
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=3).to(device)
optimizer = AdamW(model.parameters(), lr=2e-5)
```

```python
def train(model, train_loader, val_loader, epochs=3):
    lr_scheduler = get_scheduler("linear", optimizer, num_warmup_steps=0,
                                  num_training_steps=len(train_loader)*epochs)

    for epoch in range(epochs):
        model.train()
        total_loss = 0
        for batch in train_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            lr_scheduler.step()

            total_loss += loss.item()

        avg_loss = total_loss / len(train_loader)
        print(f"Epoch {epoch+1} | Training Loss: {avg_loss:.4f}")

        # Doğruluk hesaplayalım
        model.eval()
        predictions, true_labels = [], []
        with torch.no_grad():
            for batch in val_loader:
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                labels = batch['labels'].to(device)

                outputs = model(input_ids=input_ids, attention_mask=attention_mask)
                logits = outputs.logits
                preds = torch.argmax(logits, dim=1)

                predictions.extend(preds.cpu().numpy())
                true_labels.extend(labels.cpu().numpy())

        acc = accuracy_score(true_labels, predictions)
        print(f"Validation Accuracy: {acc:.4f}")
```

```python
class SceneDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_len=64):
        self.data = dataframe
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        text = self.data.iloc[idx]['scene']
        label = self.data.iloc[idx]['label']

        encoded = self.tokenizer(
            text,
            padding='max_length',
            truncation=True,
            max_length=self.max_len,
            return_tensors='pt'
        )

        return {
            'input_ids': encoded['input_ids'].squeeze(0),
            'attention_mask': encoded['attention_mask'].squeeze(0),
            'labels': torch.tensor(label, dtype=torch.long)
        }
```
